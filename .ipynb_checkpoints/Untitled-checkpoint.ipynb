```python
from OneVarBootstrap import *
```

```python
mu_true = 0
mu_gen = 0.5
smearing = 0.2
N = int(1e6)
n_bootstraps = 10**1
ibu_iterations = 12

data_streams = generate_data(mu_true, mu_gen, smearing, N)
```

```python
bins, n_bins, bin_centers, bin_widths = create_bins(data_streams[0], data_streams[1], 0.75)
```

```python
bootstrapped_data = bootstrap_data(N, n_bootstraps, data_streams)
```

```python
digitized_data, response_matrices, normalized_matrices = create_response_matrices(bootstrapped_data,
                                                                                  bins, n_bootstraps, n_bins)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 6))  # Two plots in one row

H_normalized = normalized_matrices[0, 1]
H_normalized_alt = normalized_matrices[1, 1]
# First response matrix
im = axs[0].imshow(H_normalized, origin='lower', cmap="Reds", vmin=0, vmax=1)
for i in range(n_bins):
    for j in range(n_bins):
        text = axs[0].text(j, i, f"{H_normalized.T[i, j]:0.2f}",
                           ha="center", va="center", color="w", fontweight="bold", fontsize=12)
axs[0].set_title("Response Matrix for Truth vs. Data")
axs[0].set_xlabel("Truth bin")
axs[0].set_ylabel("Data bin")
fig.colorbar(im, ax=axs[0], label='Pr(Data | Truth)')

# Second response matrix
im_alt = axs[1].imshow(H_normalized_alt, origin='lower', cmap="Blues", vmin=0, vmax=1)
for i in range(n_bins):
    for j in range(n_bins):
        text = axs[1].text(j, i, f"{H_normalized_alt.T[i, j]:0.2f}",
                           ha="center", va="center", color="w", fontweight="bold", fontsize=12)
axs[1].set_title("Response Matrix for Gen vs. Sim")
axs[1].set_xlabel("Gen bin")
axs[1].set_ylabel("Sim bin")
fig.colorbar(im_alt, ax=axs[1], label='Pr(Sim | Gen)')

plt.tight_layout()
plt.show()
```

```python
marginals = compute_marginals(response_matrices, n_bootstraps, n_bins)
```

```python
ibu_results, ibu_normalized = generate_ibu_results(n_bootstraps, n_bins, 
                                                   response_matrices, marginals, 
                                                   ibu_iterations, bin_widths)
```
